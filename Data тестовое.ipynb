```python
# Delivery rate - это процент писем, которые дошли до адресатов email-рассылки, среди всех отправленных сообщений. 

delivery_rate = round(data['Доставлено'].sum() / data['Отправлено'].sum()*100, 2)

print(f'Delivery rate = {delivery_rate}%')
```

```python
# Open rate — процент писем, которые открыли получатели, среди всех доставленных сообщений.

open_rate = round(data['Открытия'].sum() / data['Доставлено'].sum()*100, 2)

print(f'Open rate = {open_rate}%')
```

```python
# Click to Open rate - CTOR — это метрика, которая показывает процентное соотношение кликов по ссылке в письме к количеству открытий

click_to_open_rate = round(data['Клики'].sum() / data['Открытия'].sum()*100, 2)

print(f'Click to Open rate = {click_to_open_rate}%')
```

```python
# Unsubscribe rate — процент отписавшихся пользователей среди тех, кто получил рассылку.

unsubscribe_rate = round(data['Отписки'].sum() / data['Доставлено'].sum()*100, 2)

print(f'Unsubscribe rate = {unsubscribe_rate}%')
```

```python
# Выяви лучшую тему
# предположу, что лучшая тема - эта та, у которой открыли письмо чаще всего, т. е. найдем тему, у которой самый большой open rate 

# добавим колонку Open rate для каждой строки
data['Open rate'] = round(data['Открытия'] / data['Доставлено']*100, 2)

# отфильтруем строки, у которых максимальный open rate и запишем темы этих строк через запятую
best_themes = data[data['Open rate'] == data['Open rate'].max()]['Тема письма '].str.cat(sep=', ')

print(f'Лучшие темы две: {best_themes}')
```

```python
# И лучший день недели (для планирования рассылок)
# предположу, что лучший день недели - тот, у которого open rate самый большой

# построим новую таблицу день недели - open rate
weekday_open_rate = data.groupby('День недели.1').agg({'Open rate': 'sum'}).reset_index()

# отфильтруем день недели с максимальным open rate 
best_weekday = weekday_open_rate[weekday_open_rate['Open rate'] == weekday_open_rate['Open rate'].max()]['День недели.1'].str.cat(sep=', ')

print(f'Лучший день недели: {best_weekday}')
```
